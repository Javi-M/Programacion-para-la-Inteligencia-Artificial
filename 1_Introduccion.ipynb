## <a id='toc1_1_'></a>[Tensores en PyTorch](#toc0_)

Tal y como reza la documentación oficial:

> Tensors are a specialized data structure that are very similar to arrays and matrices. 
> In PyTorch, we use tensors to encode the **inputs** and **outputs** of a model, as well as 
> the model’s parameters.
> Tensors are similar to NumPy’s ndarrays, except that tensors can run on GPUs 
> or other specialized hardware to accelerate computing. If you’re familiar with 
> ndarrays, you’ll be right at home with the Tensor API. If not, follow along in
> this quick API walkthrough.
https://docs.pytorch.org/tutorials/beginner/blitz/tensor_tutorial.html

Son arrays multidimensionales. Pueden estar en la GPU. Se utilizan en PyTorch para:
- Inputs. (Es decir: cada input de una red neuronal **deberá ser un tensor**).
- Outputs
- Parámetros

Son de la clase `torch.Tensor`.
